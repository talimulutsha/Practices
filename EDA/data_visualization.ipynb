# data viz

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import online data forom online seaborn libraries 'titanic'
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='#00FF00')#, figsize=(10, 5))
```

```python
# using groupby function to sex column according to age
df.groupby('sex')[['age']].mean().plot(kind='bar', color='#00FF00')
#adding label in x axis
plt.xlabel('gender')
#adding label in y axis
plt.ylabel('mean age')
#plt.show()
```

```python
# seaborn plot
# barplot using seaborn
sns.barplot(data = df, x='survived', y='fare' )
```

```python
# using seaborn library to show bar plot and on x axis will be sex, y axis will be fare and hue will be class
sns.barplot(data = df, x='sex', y='fare', hue='class')
```
