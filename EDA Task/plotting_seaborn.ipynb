# Seaborn and plotting

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Importing things for plotting according to grammer of graphics

2. Types of plot
3. x axis
4. y axis
5. Grouping (hue/color) based on categorical varriable

7. Color palletes of the plot

```python
# load dataset form seaborn
df = sns.load_dataset('titanic')
```

```python
# scatter plot
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='age', y='fare', data=df, hue='sex')
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex', palette={'male': 'black', 'female': 'brown'})
```

```python
plt.title('Scatter plot of age vs fare by sex')
plt.xlabel('Age')
plt.ylabel('Fare')
# saving the plot with 300dpi
plt.savefig('scatter_plot.png', dpi=300)
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex', palette={'male': 'red', 'female': 'blue'})

plt.title('Scatter plot of age vs fare by sex')
plt.xlabel('Age')
plt.ylabel('Fare')
# saving the plot with 300dpi
plt.savefig('scatter_plot.png', dpi=300)

```

Interpertation:
1. We have Age of passengers at x-axis and fare at y-axis
2. The scatter plot shows the relationship between age and fare of titanic passengers.
3. We further groupd the passengers by there genders such as male(red points) and female(blue points)
4. Two male passengers gave the highest fare in titanic, and they might be travelling first class
5. The oldest person at the titanic is also a male having 80 years of age as shown in the scatter plot.

# Box plot

```python
df['class'].unique()
```

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(12,6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box plot of Age, based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')
```

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(12,6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```

# Bar plot

```python
# bar plot using seaborn

# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(12,6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette, capsize=0.1)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```

```python
# bar plot using seaborn

# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(12,6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.violinplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```
